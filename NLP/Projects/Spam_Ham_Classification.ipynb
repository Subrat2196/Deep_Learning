```python
import pandas as pd
```

```python
msg = pd.read_csv('spam.csv',encoding="ISO-8859-1")
```

```python
msg=msg.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
msg.columns = ['label','message']
```

## Data Cleaning and Preprocessing

#### Step 1 -> Remove Special Characters uisng regular expression
#### Step 2 -> Lower casing the messages
#### Step 3 -> remove stopwords
#### Step 4 -> we will make the words from sentences using split or word_tokenize
#### Step 5 -> we will lemmatize each word using list comprehension
#### Step 6 -> we will join them back

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk import WordNetLemmatizer
```

```python
lm = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len(msg)):
    review = re.sub(r'[^a-zA-Z]',' ',msg['message'][i])
    review = review.lower()
    review = review.split()
    review = [lm.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create Bag of Words 

#### bag of words can be implemented using Count vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

#### create cv instance

```python
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

### cv.fit_transform

#### The fit_transform method combines two steps: 
#### Fit: It learns the vocabulary of the text data. In other words, it finds all the unique words in your text documents and assigns each word a unique index. 
#### Transform: It converts the text documents into a matrix where each row represents a document and each column represents a word from the vocabulary. The values in the matrix are the counts of each word in the respective document.

### Here X is the independent feature or input feature

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
X.shape
```

```python
cv.vocabulary_
```

## Output feature -> label

##### pd.get_dummies() -> pd.get_dummies is a function in pandas that converts categorical variables into a form that can be provided to machine learning algorithms to do a better job in prediction. This is also known as one-hot encoding

##### One-Hot Encoding: A process of converting categorical data variables so they can be provided to ML algorithms to improve predictions. In this process, each category is converted into a new column, and a binary value (0 or 1) is assigned to each column

```python
Y = pd.get_dummies(messages['label'])
```

```python
Y
```

#### Here any one of the two can be used as output feature

```python
Y = Y.iloc[:,0].values   # Here iloc selects the first column in the pandas dataframe and creates a series , .values is used to convert it to numpy array
```

```python
Y
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
```

### Importing naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB
spam_detec_model = MultinomialNB()
spam_detec_model.fit(X_train,Y_train)
```

```python
spam_detec_model.predict(X_test)
```

```python
Y_Pred = spam_detec_model.predict(X_test)
```

```python
Y_Pred
```

#### Performance Matrix -> Accuracy score and classification report

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy = accuracy_score(Y_test,Y_Pred)
accuracy
```

```python
classif_score = classification_report(Y_test,Y_Pred)
print(classif_score)
```
