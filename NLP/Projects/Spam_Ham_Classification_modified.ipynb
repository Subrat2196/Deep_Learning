## We need to reorder the BOW and TF-IDF steps with the train-test split to prevent data leakage

```python
import pandas as pd
```

```python
message = pd.read_csv('spam.csv',encoding="ISO-8859-1")
```

### Step 1 -> Preprocessing and Data Cleaning

```python
from nltk.corpus import stopwords
```

```python
from nltk import WordNetLemmatizer
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
message.head()
```

```python
message = message.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
message.columns=['label','content']
```

```python
message
```

#### here we will remove the special character, convert to lowercase , sentence to array of words,lemmatize words , join them back 

```python
import re 
lemma = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(message)):
    review = re.sub('[^a-zA-Z]',' ',message['content'][i])
    review = review.lower()
    review = review.split()
    review = [lemma.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

### output feature 

```python
message
```

```python
y=pd.get_dummies(message['label'])
```

```python
y = y.iloc[:,0].values
```

```python
y
```

## Train-Test Split -> Updated step before BOW / TF-IDF

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(corpus,y,test_size=0.2)
```

### Bag of Words -> fit_transform and transform

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,3))
```

```python
X_train = cv.fit_transform(X_train).toarray()   # Here we did it on X_train since we need that to convert into vectors
X_test = cv.transform(X_test).toarray() # Similar to linear regression 
```

```python
cv.vocabulary_
```

## Model  Apply

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect = MultinomialNB()
```

```python
spam_detect.fit(X_train,Y_train)
```

```python
Y_pred = spam_detect.predict(X_test)
Y_pred
```

### Performance Metrics 

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
acc = accuracy_score(Y_test,Y_pred)
```

```python
acc
```

```python
cr = classification_report(Y_test,Y_pred)
print(cr)
```
